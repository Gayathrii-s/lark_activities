```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_excel('Telco-Customer-Churn.xlsx')
```

```python
data
```

```python
data.head()
```

#             Carry out Visualization for the following questions

# Compare churn count with respect to gender

```python
plt.figure(figsize=(10,8))
sns.countplot(data['gender'],hue=data['Churn'])
plt.title('Plot of Churn count VS Gender',fontsize=20,weight='bold',c='blue')
plt.xlabel('Gender',fontsize=20)
plt.ylabel('Count',fontsize=20)
plt.grid(linestyle= '--')
plt.show()
```

# Find out how many female senior citizens there in the dataset

```python
F=data[(data['gender']=='Female') & (data['SeniorCitizen']==1)]
F                                    
```

```python
F['gender'].count()
```

```python
plt.figure(figsize=(10,8))
sns.countplot(data['gender'],hue=data['SeniorCitizen'],data=data)
plt.xlabel('gender',fontsize=18)
plt.ylabel('Count',fontsize=18)
plt.grid(linestyle ='--')
plt.show()
```

# Compare 'tenure' with 'Total Charges'

```python
plt.figure(figsize=(15,12))
plt.scatter(data['tenure'],data['TotalCharges'],s=50,c='red',marker="*")
plt.title('Plot of Tenure VS TotalCharges ',fontsize=30,fontstyle='oblique',weight='bold')
plt.xlabel('Tenure',fontsize=15)
plt.ylabel('TotalCharges',fontsize=15)
plt.show()
```

# Find out which contract is preferred by the senior citizen

```python
S=data[data['SeniorCitizen']==1][['SeniorCitizen','Contract']]
S
```

```python
plt.figure(figsize=(12,10))
sns.countplot(data['Contract'],hue=data['SeniorCitizen'])
plt.title('Plot of Contract VS SeniorCitizen',fontsize=25,fontstyle='normal',weight='bold')
plt.xlabel('Contract',fontsize=15)
plt.ylabel('Count',fontsize=15)
plt.grid(linestyle= '--')
plt.show()
```
