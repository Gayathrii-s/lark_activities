# Examine the data by using visualizations.

```python
plt.figure(figsize=(12,10))
sns.boxplot(x='Region',y='GDP per capita (current US$)',data=data)
plt.title('Boxplot of Region  VS GDP',fontsize=15,fontstyle='oblique',weight='bold')
plt.xlabel('Region',fontsize=15)
plt.ylabel('GDP per capita (current US$)',fontsize=15)
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.barplot(x='country',y='Population in thousands (2017)',data=data.nlargest(15,'Population in thousands (2017)'))
plt.title('Barplot of country  VS Population',fontsize=25,fontstyle='oblique',weight='bold')
plt.xlabel('country',fontsize=15)
plt.ylabel('Population in thousands (2017)',fontsize=15)
plt.xticks(rotation=90)
plt.grid(linestyle='--')
plt.show()
```

```python
data[['GDP per capita (current US$)']].plot.kde()
```

```python
slice_data=data[['Surface area (km2)','GDP per capita (current US$)','Population in thousands (2017)']]
```

```python
sns.heatmap(slice_data.corr(),annot=True)
plt.title('Heatmap')
plt.xticks(rotation=90)
```

```python
plt.figure(figsize=(10,8))
sns.pairplot(data)
plt.show()
```
